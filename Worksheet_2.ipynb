```python
#  Problem 2- Data Imputations:
#  1. Load the provided dataset and import in pandas DataFrame.

medical=pd.read_csv("medical_students_dataset.csv")
print("Medical Records:")
medical.head()
```

```python
#2. Check info of the DataFrame and identify column with missing (null) values.

print("Overview of Medical Records:")
medical.info()
print("Columns with null values:")
sn=1
columns_with_null_values_check=medical.isnull().sum()
for columns,null_count in columns_with_null_values_check.items():
  if null_count>0:
    print(f"{sn}. {columns}")
    sn+=1

```

```python
# 3. For the column with missing values fill the values using various techniques we discussed above. Try to explain why did you select the particular methods for particular column.

for column in medical.columns:
  if column=="Student ID":
    medical[column]=medical.index+1
    continue
  if medical[column].dtype=="float64":
    medical[column]=medical[column].fillna(medical[column].mean())
  else:
    medical[column]=medical[column].fillna(medical[column].mode()[0])

print("Filled up Medical Reports:")
medical.head()
```

```python
#4. Check for any duplicate values present in Dataset and do necessary to manage the duplicate items.{Hint: dataset.duplicated.sum()}

if medical.duplicated().sum()==0:
  print("There are no duplicated values.")
else:
  print("These values are duplicated:\n",medical[medical.duplicated()])
```

```python
# 3.2 Exercises- Data Cleaning and Transformations with ”Titanic Dataset”:
#Problem- 1:

titanic=pd.read_csv("Titanic-Dataset.csv")
titanic_passengers=titanic[["Name","Pclass","Sex","Age","Fare","Survived"]]
print("First Five Titanic Passengers Information:")
display(titanic_passengers.head())

first_class_passengers_only=titanic_passengers[titanic_passengers["Pclass"]==1]
print("\nSome of the first class Passengers Information:")
display(first_class_passengers_only.head())

fare_mean=first_class_passengers_only["Fare"].mean()
fare_median=first_class_passengers_only["Fare"].median()
fare_max=first_class_passengers_only["Fare"].max()
fare_min=first_class_passengers_only["Fare"].min()
print(f"\nAverage fare of first class passengers : {fare_mean}\nMedian fare of first class passengers : {fare_median}")
print(f"\nHighest fare of first class passengers : {fare_max}\nLowest fare of first class passengers : {fare_min}")


```

```python
#Problem- 2:

print("The shape of the original dataset is ",titanic_passengers.shape)
null_count=titanic_passengers["Age"].isnull().sum()
print(f"There are {null_count} null values in the Age column.")
null_removed_titanic_passengers=titanic_passengers.dropna(subset=["Age"])
print("After dropping null values in Age column,the shape of the dataset is:",null_removed_titanic_passengers.shape)
```

```python
#Problem- 3:

print("Original dataset:")
display(titanic.head())
#1. Use one-hot encoding to convert the ’Embarked’ column into separate binary columns (’Embarked C’,’Embarked Q’, ’Embarked S’).
encoded_embarked_columns=pd.get_dummies(titanic["Embarked"],prefix="Embarked")

#2. Add these new columns to the original DataFrame.
titanic=pd.concat([titanic,encoded_embarked_columns],axis=1)

#3. Drop the original ’Embarked’ column.
new_titanic=titanic.drop("Embarked",axis=1)

#4. Print the first few rows of the modified DataFrame to verify the changes.
print("New dataset:")
new_titanic.head()
```

```python
#Problem- 4:

import matplotlib.pyplot as plt

titanic_passengers_by_sex=titanic[["Sex","Survived"]]
avg_survival_rate_by_sex=titanic_passengers_by_sex.groupby("Sex")["Survived"].mean().reset_index()
avg_survival_rate_by_sex.columns=["Sex","Survival Rate(%)"]
avg_survival_rate_by_sex["Survival Rate(%)"]=(avg_survival_rate_by_sex["Survival Rate(%)"]*100).round(2)
display(avg_survival_rate_by_sex)

avg_survival_rate_by_sex.plot(kind="bar",x="Sex",y="Survival Rate(%)",legend=False,color=["pink","blue"])
plt.title("Average survival rate by sex")
plt.xlabel("Sex")
plt.ylabel("Survival Rate")
plt.ylim(0,100)
plt.grid()
plt.show()
```

```python
#Problem- 5:

people_from_each_port=titanic["Embarked"].value_counts().reset_index()
people_from_each_port.columns=["Port","Number of Passengers"]
display(people_from_each_port)

people_from_each_port.plot(x="Port",y="Number of Passengers",kind="bar",color=["red","green","blue"],legend=False)
plt.title("Number of passengers from each port")
plt.xlabel("’S’ (Southampton)        ’C’ (Cherbourg)              ’Q’ (Queenstown)")
plt.ylabel("Number of Passengers")
```
