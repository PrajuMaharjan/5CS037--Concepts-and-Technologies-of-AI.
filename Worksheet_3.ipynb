```python
#3.To-Do-Task
#3.1 Warm Up Exercises:
#1. Sorting and Subsetting:
#1. Load the provided dataset and import in pandas DataFrame.

import pandas as pd
read_csv=pd.read_csv("Titanic-Dataset.csv")
titanic=pd.DataFrame(read_csv)

#2. Check info of the DataFrame and identify following:

print("Info of titanic.csv : ")
titanic.info()
```

```python
#Problem 1- Sorting:
#1. Create a DataFrame called fare that contains only the Fare column of the Titanic dataset. Print the head of the result.

fare=titanic["Fare"]
print("Fare dataset:")
fare.head()
```

```python
#2. Create a DataFrame called class age that contains only the Pclass and Age columns of the Titanic dataset, in that order. Print the head of the result.

class_age=titanic[["Pclass","Age"]]
print("Passenger Class and Age columns:")
class_age.head()
```

```python
#3. Create a DataFrame called survived gender that contains the Survived and Sex columns of the Titanic dataset, in that order.Print the head of the result.

survived_gender=titanic[["Survived","Sex"]]
print("Survived and Sex columns:")
survived_gender.head()
```

```python
#Problem- 2- Subsetting:
#Subsetting Rows:
#1.	Filter the Titanic dataset for cases where the passenger’s fare is greater than 100, assigning it to fare_gt 100.View the printed result.

fare_gt_100=titanic[titanic["Fare"]>100]
print("First 10 passengers with fare greater than 100:")
fare_gt_100.head(10)
```

```python
#2.Filter the Titanic dataset for cases where the passenger’s class (Pclass) is 1, assigning it to first class.View the printed result.

first_class=titanic[titanic["Pclass"]==1]
print("5 first class passengers : ")
first_class.head()
```

```python
#3.	Filter the Titanic dataset for cases where the passenger’s age is less than 18 and the passenger is female (Sex is "female"), assigning it to female under 18. View the printed result.

female_under_18=titanic[(titanic["Age"]<18) & (titanic["Sex"]=="female")]
print("Information of 5 females under 18 on the Titanic:")
female_under_18.head()
```

```python
#Subsetting Rows by Categorical variables:
#1.Filter the Titanic dataset for passengers whose Embarked port is either "C" (Cherbourg) or "S" (Southampton), assigning the result to embarked c  or s. View the printed result.

embarked_c_or_s=titanic[(titanic["Embarked"]=="C")| (titanic["Embarked"]=="S")]
print("5 passengers who embarked from Southampton or Cherbourg : ")
embarked_c_or_s.head()
```

```python
#2.Filter the Titanic dataset for passengers whose Pclass is in the list [1, 2] (indicating first or second class), assigning the result to first second class.View the printed result.

first_second_class=titanic[(titanic["Pclass"]==1) | (titanic["Pclass"]==2)]
print("Information of five first or second class passengers : ")
first_second_class.head()
```

```python
# Which passenger had the highest fare paid relative to their age?
# To answer the question perform following operations:

# 1.	Add a column to the Titanic dataset, fare per  year, containing the fare divided by the age of the passenger(i.e.,Fare/Age).
titanic["fare_per_year"]=titanic["Fare"]/titanic["Age"]

# 2.	Subset rows where fare per year is higher than 5, assigning this to high fare  age.
high_fare_age=titanic[titanic["fare_per_year"]>5]

# 3.	Sort high fare  age by descending fare  per year, assigning this to high fare age srt.
high_fare_age_srt=high_fare_age.sort_values(by="fare_per_year",ascending=False)

# 4.	Select only the Name and fare  per year columns of high fare age srt and save the result as result.
result=high_fare_age_srt[["Name","fare_per_year"]]

# 5.	Look at the result.
print("the top five passengers who had the highest fare paid relative to their age : ")
display(result.head())
```

```python
# Which adult male passenger (age ≥ 18 and Sex is ’male’) paid the highest fare relative to their class?
# To answer the question perform following operations:
# 1.	Add a column to the Titanic dataset, fare per class, containing the fare divided by the passenger class i.e. Fare / Pclass.
titanic["fare_per_class"]=titanic["Fare"]/titanic["Age"]

# 2.	Subset rows where the passenger is male (Sex is ”male”) and an adult (Age is greater than or equal to 18), assigning this to adult males.
adult_males=titanic[(titanic["Sex"]=="male") &(titanic["Age"]>18)]

# 3.	Sort adult males by descending fare per class, assigning this to adult males srt.
adult_males_srt=adult_males.sort_values(by="fare_per_class",ascending=False)

# 4.	Select only the Name, Age, and fare per  class columns of adult  males  sr and save the result as result.
result=adult_males_srt[["Name","Age","fare_per_class"]]

# 5.	Look at the result.
print("Top 5 adult male passengers who paid the highest fare relative to their class : ")
result.head()
```

```python
#3.3	Exploratory Data Analysis with Group-by Method Practice Exercise:
# What percent of the total fare revenue came from each passenger class?

# 1.	Calculate the total Fare paid across all passengers in the Titanic dataset.
total_fare_paid=titanic["Fare"].sum()
print("Total fare paid across all classes : ",total_fare_paid)

# 2.	Subset for passengers in first class (Pclass is 1) and calculate their total fare.
first_class=titanic[titanic["Pclass"]==1]
total_fare_paid_first_class=first_class["Fare"].sum()
print("Total fare paid by first class passengers : ",total_fare_paid_first_class)

# 3.	Do the same for second class (Pclass is 2) and third class (Pclass is 3).
second_class=titanic[titanic["Pclass"]==2]
total_fare_paid_second_class=second_class["Fare"].sum()
print("Total fare paid by second class passengers : ",total_fare_paid_second_class)

third_class=titanic[titanic["Pclass"]==3]
total_fare_paid_third_class=third_class["Fare"].sum()
print("Total fare paid by third class passengers : ",total_fare_paid_third_class)

# 4.	Combine the fare totals from first, second, and third classes into a list.
fare_totals=[total_fare_paid_first_class,total_fare_paid_second_class,total_fare_paid_third_class]
print("")

# 5.	Divide the totals for each class by the overall total fare to get the proportion of fare revenue by class.
print(f"First class passengers paid {(fare_totals[0]/total_fare_paid)*100} of the total fare revenue.")
print(f"Second class passengers paid {(fare_totals[1]/total_fare_paid)*100} of the total fare revenue.")
print(f"Third class passengers paid {(fare_totals[2]/total_fare_paid)*100} of the total fare revenue.")
```

```python
#  What percent of the total number of passengers on the Titanic belonged to each age group(e.g., child, adult, senior)?

# 1.	Create a new column, age group, that categorizes passengers into "child" (age < 18), "adult" (age 18{64), and "senior" (age 65 and above).
age_groups=[]
for age in titanic["Age"]:
  if pd.isna(age):
    age_groups.append("Unknown")
  elif age<18:
    age_groups.append("Child")
  elif age<=64:
    age_groups.append("Adult")
  else:
    age_groups.append("Senior")
titanic["age_group"]=age_groups

# # 2. Calculate the total number of passengers on the Titanic.
print("Total number of passengers on the Titanic : ",len(titanic))

# 3.	Count the number of passengers in each age group.
age_groups_number=titanic.groupby("age_group").size()
print("Number of passengers in each age group : ")
display(age_groups_number)

# 4.	Divide the count of each age group by the total number of passengers to get the proportion of passengers in each age group.
age_groups_proportion=(age_groups_number/len(titanic))

# 5.	Display the proportion as a percentage.
age_groups_percentage=age_groups_proportion*100
print("The percentage of total number of passengers on the Titanic belonging to each age group : ")
display(age_groups_percentage)
```
