```python
#Problem- 4: Matrix Operations:

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

#1. Prove A.A−1 = I.

A_inv=np.linalg.inv(A)
dot=np.round(np.dot(A,A_inv))
id_matrix=np.identity(2)

print("A.A−1 = ",np.round(dot,0))
print("Identity matrix = ",id_matrix)
if np.allclose(dot,id_matrix):
  print("A.A−1 = I \n Hence,proved")
```

```python
#2. Prove AB̸ = BA.

AB=np.dot(A,B)
BA=np.dot(B,A)
print("AB = ",AB,"\n BA = ",BA)
if np.allclose(AB,BA):
  print("AB = BA.\n Hence proved")
else:
  print("AB =/= BA.\n Hence proved")
```

```python
#3. Prove (AB)T = BTAT.

AB_T=AB.T
B_T=B.T
A_T=A.T
B_T_A_T=np.dot(B_T,A_T)

print("ABT = ",AB_T)
print("BTAT = ",B_T_A_T)
if (np.allclose(AB_T,B_T_A_T)):
  print("(AB)T = BTAT.\n Hence proved.")
```

```python
# Solve the following system of Linear equation using Inverse Methods.
#  2x −3y+z =−1
#  x −y+2z =−3
#  3x +y−z =9

A=np.array([[2,-3,1],[1,-1,2],[3,1,-1]])
B=np.array([-1,-3,9])
X=(["x","y","z"])
A_inverse=np.linalg.inv(A)
Solution=np.dot(A_inverse,B)
print("2x −3y+z =−1\nx −y+2z =−3\n3x +y−z =9")
print("Solution:\n")
for var,ans in zip(X,Solution):
  print(var," = ",ans)
```

```python
#1. Element-wise Addition:

import numpy as np
import time

list_1=list(range(1000000))
list_2=list(range(1000000))
sum_1=[]
start_time=time.time()
for a,b in zip(list_1,list_2):
        sum_1.append(a+b)
end_time=time.time()
time_taken=end_time-start_time
print("Time taken to perform element-wise addition using python lists:",time_taken,"seconds.")

arr_1=np.arange(1000000)
arr_2=np.arange(1000000)

start_time=time.time()
sum_2=arr_1+arr_2
end_time=time.time()
time_taken=end_time-start_time
print("Time taken to perform element-wise addition using numpy array:",time_taken,"seconds.")
```

```python
#Element wise Multiplication

list_1=list(range(1000000))
list_2=list(range(1000000))
product_1=[]
start_time=time.time()
for a,b in zip(list_1,list_2):
        product_1.append(a*b)
end_time=time.time()
time_taken=end_time-start_time
print("Time taken to perform element-wise multiplication using python lists:",time_taken,"seconds.")

arr_1=np.arange(1000000)
arr_2=np.arange(1000000)

start_time=time.time()
pro_2=arr_1*arr_2
end_time=time.time()
time_taken=end_time-start_time
print("Time taken to perform element-wise multiplication using numpy array:",time_taken,"seconds.")
```

```python
#3. Dot Product

list_1=list(range(1000000))
list_2=list(range(1000000))
start_time=time.time()
product_1=[]
for a,b in zip(list_1,list_2):
        product_1.append(a*b)

dot_product_1=sum(product_1)
end_time=time.time()
time_taken=end_time-start_time
print("Time taken to compute dot product using python lists:",time_taken,"seconds.")

arr_1=np.arange(1000000)
arr_2=np.arange(1000000)

start_time=time.time()
dot_pro_2=np.dot(arr_1,arr_2)
end_time=time.time()
time_taken=end_time-start_time
print("Time taken to compute dot product using numpy array:",time_taken,"seconds.")
```
