```python
# 5. Split your data into Feature(X) and Label(Y).

X=student[["Math","Reading"]]
Y=student["Writing"]
display(X.head())
print("\n")
display(Y.head())
```

```python
# To-Do-2:
# 1. To make the task easier- letâ€™s assume there is no bias or intercept.
# 2. Create the following matrices:
# Y= WT X

import numpy as np

X_matrix=X.values.T #Convert dataset to matrix
W=np.array([[0.6],[0.4]]) #Weight vector defined
W_T=W.T
Y_matrix=W_T @ X_matrix

print("Matrix X : ",X_matrix.shape)
print("Matrix W^T : ",W_T.shape)
print("Matrix Y : ",Y_matrix.shape)
#Matrix itself not printed because of large number of values
```

```python
# To-Do-3:
# 1. Split the dataset into training and test sets.You can use an 80-20 orb 70-30 split,with 80% (or70%) of the data used for training and the rest for testing.

def train_test_split(x,y,test_size=0.2,random_seed=42):
  np.random.seed(random_seed)
  indices=np.arange(x.shape[0])
  np.random.shuffle(indices)
  test_split_size = int(len(x) * test_size)
  test_indices=indices[:test_split_size]
  train_indices=indices[test_split_size:]
  X_train,X_test=x[train_indices], x[test_indices]
  y_train,y_test=y[train_indices], y[test_indices]
  return X_train,X_test,y_train,y_test

X_arr=X.values
Y_arr=Y.values
X_train,X_test,y_train,y_test=train_test_split(X_arr,Y_arr,test_size=0.2)

print("X_train:\n", X_train.shape)
print("X_test:\n", X_test.shape)
print("y_train:\n", y_train.shape)
print("y_test:\n", y_test.shape)
```

```python
# To - Do - 4: Building a Cost Function
def cost_function(x,y,w):
  y_pred= x @ w; #Predictions
  error=y_pred -y
  cost=(1/x.shape[0])*np.sum(error**2)
  return cost

#Test Case
A=np.array([[1,2],[3,4],[5,6]])
B=np.array([[3],[7],[11]])
C=np.array([[1],[1]])
cost=cost_function(A,B,C)
print("Mean Squared Error : ",cost)

#Tested using student dataset
y_arr=Y_arr.reshape(-1,1)#reshaped to make it 2d array
cost_2=cost_function(X_arr,y_arr,W)
print("Mean Squared Error : ",cost_2)

```

```python
# To - Do - 6:Gradient Descent from Scratch:

def gradient_descent(X, Y, W, alpha, iterations=1001):
  for i in range(iterations):
    y_pred=X @ W
    error=y_pred-Y
    gradient=(2/X.shape[0])*(X.T @error)
    W=W-alpha*gradient
    if i % 100==0:
      cost=(1/X.shape[0])*np.sum(error**2)
      print(f"Iteration no. {i} : Cost : {cost}")#prints cost every 100 iterations

  return W #Returns final optimized parameters

#Test case
X_gd = np.array([[1,2],[3,4],[5,6]])
Y_gd = np.array([[3],[7],[11]])
W_init = np.array([[0],[0]])#Initial parameters=0,0

W_opt = gradient_descent(X_gd, Y_gd, W_init,0.001)
print("Optimal weights:\n", W_opt)

#Test using student dataset
print("Testing using student dataset : ")
W_opt_2=gradient_descent(X_arr,y_arr,W,0.0001)
print("Optimal Weights : \n",cost_2)
```

```python
# To - Do - 8:Implementation of RMSE in the Code

def rmse(y_true,y_pred):
  error=y_pred-y_true
  mse=(1/y_true.shape[0])*np.sum(error**2)
  return np.sqrt(mse)

# Using weights from gradient descent function
y_pred = X_gd @ W_opt
error_rmse = rmse(Y_gd,y_pred)
print("RMSE:", error_rmse)
```

```python
# To- Do- 9-Code for R-Squared Error:

def r2(Y,Y_pred):
    res_ss=np.sum((Y-Y_pred)**2)#Residual sum of squares
    total_ss=np.sum((Y-np.mean(Y))**2)#Total sum of squares
    r_2=1-(res_ss/total_ss)
    return r_2

# Using values from gradient descent function
y_pred= X_gd @ W_opt
r_2 =r2(Y_gd, y_pred)
print("R2 : ", r_2)
```

```python
# To - Do - 10:Compiling everything:

def main():
  data=pd.read_csv("student.csv")

  x=data[["Math","Reading"]].values
  y=data["Writing"].values
  y=y.reshape(-1,1)
  X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_seed=42)
  W=np.zeros((X_train.shape[1],1))
  alpha=0.00001
  iterations=1000
  W_optimal= gradient_descent(X_train, Y_train, W, alpha, iterations)
  Y_pred = np.dot(X_test, W_optimal)
  model_rmse = rmse(Y_test, Y_pred)
  model_r2 = r2(Y_test, Y_pred)
  print("Final Weights:", W_optimal)
  print("RMSE on Test Set:", model_rmse)
  print("R-Squared on Test Set:", model_r2)

if __name__=="__main__":
  main()
```
